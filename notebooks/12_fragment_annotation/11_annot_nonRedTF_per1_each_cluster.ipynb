```bash
### set environment
source ../config_duke.sh
source ../config_load_module_bedtools.sh

### run script using sbatch
sbatch -p scavenger \
    --array=0-15 \
    --mem 8G \
    -o ${FD_LOG}/annot_fragment_per1_each_cluster.%a.txt \
    <<'EOF'
#!/bin/bash
### set directories and global parameters
source ../config_duke.sh
CHROM=chr17
TARGET=target_PER1

### set input and output files
# input count files
FD_CNTS=($(ls -d ${FD_WRK}/count_fragment/*/))
FD_CNT=${FD_CNTS[${SLURM_ARRAY_TASK_ID}]}
FN_CNT=${TARGET}.bed

# motif annotation
FD_MTF=${FD_ANN}/motif_cluster_jvierstra/hg38_archetype_motifs_v1/${CHROM}

# genome annotation
FD_GEN=${FD_ANN}/genome/hg38/fasta
FN_GEN=${CHROM}.fa

### print start message
echo "Slurm Array Index: " ${SLURM_ARRAY_TASK_ID}
echo "Input  file:       " ${FD_CNT}/${FN_CNT}
echo
echo "Show the first few lines of the input file"
echo ${FD_CNT}/${FN_CNT}
head ${FD_CNT}/${FN_CNT}

### annotation fragments
for FP_MTF in $(ls ${FD_MTF}/*_merge.bed); do
    ### get motif name
    FNAME=$(basename ${FP_MTF})
    MOTIF="${FNAME%_*.*}"
    
    ### set output folder and file name
    SAMPLE=$(basename ${FD_CNT})
    FD_OUT=${FD_WRK}/annotation_fragment/${SAMPLE}/${MOTIF}
    FN_OUT=${TARGET}.bed
    FN_NUC=${TARGET}_nuc.bed
    
    ### annotation using intersect
    echo "Motif  file: " ${FP_MTF}
    echo "Output file: " ${FD_OUT}/${FN_OUT}
    echo "Output file: " ${FD_OUT}/${FN_NUC}
    
    mkdir -p ${FD_OUT}
    bedtools intersect -a ${FD_CNT}/${FN_CNT} -b ${FP_MTF} -wo    > ${FD_OUT}/${FN_OUT}
    bedtools nuc -fi ${FD_GEN}/${FN_GEN} -bed ${FD_OUT}/${FN_OUT} > ${FD_OUT}/${FN_NUC}
    
    ### print end message
    head -n 3 ${FD_OUT}/${FN_OUT}
    head -n 3 ${FD_OUT}/${FN_NUC}
done

echo "Done!"

EOF
```
