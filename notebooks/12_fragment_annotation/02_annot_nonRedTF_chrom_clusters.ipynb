## Annotate upstream of PER1 region

```bash
ls -d ${FD_LOG}
echo ${NODE}
```

**RUN**

```bash
### set env
source ../config_duke.sh
source ../config_load_module_bedtools.sh

sbatch -p ${NODE} \
    --array 0-23 \
    --mem 20G \
    -o ${FD_LOG}/annotate_fragment_chrom_each_clusters.%a.txt \
    <<'EOF'
#!/bin/bash
### set directories & global variables
source ../config_duke.sh -v
CHROMS=($(seq 1 22) X Y)
CHROM=chr${CHROMS[${SLURM_ARRAY_TASK_ID}]}
TARGET=${CHROM}
#CHROM=chr17
#TARGET=target_PER1

### set input and output
FD_BEDS=($(ls -d ${FD_RES}/count_fragment/{Input?*,TFX?_DMSO,TFX?_Dex}/))
FN_BED=${TARGET}.bed.gz

### motif annotation
#FD_MTF=${FD_ANN}/motif_cluster_jvierstra/hg38_archetype_motifs_v1
#FN_MTF=${CHROM}_rm_mouse_merge.bed.gz
FP_MTFS=$(ls ${FD_ANN}/motif_cluster_jvierstra/hg38_archetype_motifs_v1/${CHROM}/*_merge.bed)


### START message
timer_start=`date +%s`
echo "Slurm Array Index: " ${SLURM_ARRAY_TASK_ID}
echo "Time Stamp:        " $(date +"%m-%d-%y+%T")
echo

### loop through each sample and count each fragment of a chromosome
echo "RUN: loop through samples and motifs to annotate fragment"

for FD_BED in ${FD_BEDS[@]}; do

    ### set input bed file and output directory
    SAMPLE=$(basename ${FD_BED})
    FD_OUT=${FD_RES}/annotation_fragment/${SAMPLE}/${TARGET}
    
    ### init: create output folder if not exist
    mkdir -p ${FD_OUT}
    
    ### print message of input file
    echo ++++++++++++++++++++++++++++++++++++++++++++++++
    echo "Input file: " ${FD_BED}/${FN_BED}
    echo "Show the first few lines of the input file"
    zcat ${FD_BED}/${FN_BED} | head -n 3
    echo
    
    for FP_MTF in ${FP_MTFS[@]}; do
        ### set output file
        FN_MTF=$(basename ${FP_MTF})
        FN_OUT=${FN_MTF}.gz
        
        ### annotation using intersect
        bedtools intersect \
            -a ${FD_BED}/${FN_BED} \
            -b ${FP_MTF} \
            -wo |\
            gzip -c > ${FD_OUT}/${FN_OUT}
    
        ### print message of output file
        echo "Show the first few lines of the output file"
        echo ${FD_OUT}/${FN_OUT}
        zcat ${FD_OUT}/${FN_OUT} | head -n 3
        echo
    done
done

### END message
timer=`date +%s`
runtime=$(echo "${timer} - ${timer_start}" | bc -l)
echo 'Done!'
echo "Run Time: $(displaytime ${runtime})"

EOF
```
