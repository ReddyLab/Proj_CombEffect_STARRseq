```r
dat1 = dat %>% 
    dplyr::filter(Motif == "NR/20") %>% 
    dplyr::select(Chrom_Frag, Start_Frag, End_Frag, Count_Frag) %>%
    mutate(Motif1 = "NR20")

dat2 = dat %>% 
    dplyr::filter(Motif == "AP1/1") %>% 
    dplyr::select(Chrom_Frag, Start_Frag, End_Frag, Count_Frag) %>%
    mutate(Motif2 = "AP1")

dat_ann = full_join(
    dat1, dat2,
    by = c("Chrom_Frag", "Start_Frag", "End_Frag", "Count_Frag"))

dat_ann = dat_ann %>% replace_na(list(Motif1 = "", Motif2 = ""))
```

```r
dat  = dat_inp

dat1 = dat %>%
    dplyr::select(Chrom_Frag, Start_Frag, End_Frag, Count_Frag) %>%
    distinct

dat2 = dat %>% 
    dplyr::filter(Motif == "NR/20") %>% 
    dplyr::select(Chrom_Frag, Start_Frag, End_Frag, Count_Frag) %>%
    mutate(NR20 = "NR20")

dat3 = dat %>% 
    dplyr::filter(Motif == "AP1/1") %>% 
    dplyr::select(Chrom_Frag, Start_Frag, End_Frag, Count_Frag) %>%
    mutate(AP1_1 = "AP1")

dat = dat1 %>%
    full_join(dat2, by = c("Chrom_Frag", "Start_Frag", "End_Frag", "Count_Frag")) %>%
    full_join(dat3, by = c("Chrom_Frag", "Start_Frag", "End_Frag", "Count_Frag"))

#dat = dat %>% replace_na(list(NR20 = "Empty", AP1_1 = "Empty"))
dat$Motif = paste(dat$NR20, dat$AP1_1, sep = "|")

dat = dat %>% uncount(Count_Frag)
dat = dat %>% 
    arrange(Start_Frag) %>% 
    mutate(idx=1:nrow(dat)) %>% 
    mutate(d=End_Frag-Start_Frag)
```

```r
dat_inp = dat_ann_inp
dat_inp = dat_inp %>% 
    group_by(Chrom_Frag, Start_Frag, End_Frag, Count_Frag) %>% 
    summarize(Count_Motif = n())
dat_inp = dat_inp %>% mutate(
    Len_Frag = End_Frag - Start_Frag,
    Loc      = (End_Frag + Start_Frag)/2)
head(dat_inp)
```

```r
dat_dmso = dat_ann_dmso
dat_dmso = dat_dmso %>% 
    group_by(Chrom_Frag, Start_Frag, End_Frag, Count_Frag) %>% 
    summarize(Count_Motif = n())
dat_dmso = dat_dmso %>% mutate(
    Len_Frag = End_Frag - Start_Frag,
    Loc      = (End_Frag + Start_Frag)/2)
head(dat_dmso)
```

```r
dat_dex = dat_ann_dex
dat_dex = dat_dex %>% 
    group_by(Chrom_Frag, Start_Frag, End_Frag, Count_Frag) %>% 
    summarize(Count_Motif = n())
dat_dex = dat_dex %>% mutate(
    Len_Frag = End_Frag - Start_Frag,
    Loc      = (End_Frag + Start_Frag)/2)
head(dat_dex)
```
