```bash
sbatch -p ${NODE} \
    --mem 8G \
    -o ${FD_LOG}/extract_fragment_region_dex_GR_P300_dnase_chr17.txt \
    <<'EOF'
#!/bin/bash
### set directories & global variables
source ../config_duke.sh

CHROM=chr17

FD_CNT=${FD_RES}/count_fragment
FN_CNT=${CHROM}.bed.gz
FD_REG=${FD_RES}/region
FN_REG=region_dex_GR_P300_dnase.bed

### print start message
timer_start=`date +%s`
echo "Slurm Array Index: " ${SLURM_ARRAY_TASK_ID}
echo "Time Stamp:        " $(date +"%m-%d-%y+%T")
echo

FP_CNTS=($(ls ${FD_CNT}/*/${CHROM}.bed.gz))
for FP_BED in ${FP_CNTS[@]}; do
    
    ### set input and output file
    FD_BED=${FP_BED%/*}
    FP_IN1=${FP_BED}
    FP_IN2=${FD_REG}/${CHROM}/${FN_REG}
    FP_OUT=${FD_BED}/region_dex_GR_P300_dnase_${CHROM}.bed.gz
    
    ### show input file
    echo ++++++++++++++++++++++++++++++++++
    echo "Input: " ${FP_IN1}
    echo "show first few lines of input"
    zcat ${FP_IN1} | head -5
    echo
    echo "Input: " ${FP_IN2}
    echo "show first few lines of input"
    cat ${FP_IN2} | head -5
    echo
    
    ### intersection
    bedtools intersect -a ${FP_IN1} -b ${FP_IN2} -wo | gzip -c > ${FP_OUT}
    
    ### show output file
    echo "Output: " ${FP_OUT}
    echo "show first few lines of input"
    zcat ${FP_OUT} | head -5
    echo
done

### print end message
timer=`date +%s`
runtime=$(echo "${timer} - ${timer_start}" | bc -l)
echo
echo 'Done!'
echo "Run Time: $(displaytime ${runtime})"

EOF
```
