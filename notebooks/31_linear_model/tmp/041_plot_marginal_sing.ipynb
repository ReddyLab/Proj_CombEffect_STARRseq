```r
### Helper function
### https://win-vector.com/2014/05/30/trimming-the-fat-from-glm-models-in-r/
stripGlmLR = function(cm) {
  cm$y = c()
  cm$model = c()
  
  cm$residuals = c()
  cm$fitted.values = c()
  cm$effects = c()
  cm$qr$qr = c()  
  cm$linear.predictors = c()
  cm$weights = c()
  cm$prior.weights = c()
  cm$data = c()

  cm$family$variance = c()
  cm$family$dev.resids = c()
  cm$family$aic = c()
  cm$family$validmu = c()
  cm$family$simulate = c()
  attr(cm$terms,".Environment") = c()
  attr(cm$formula,".Environment") = c()
  
  return(cm)
}
```

```r
fun = function(motifs){
    res = sapply(motifs, function(motif){
        if (motif == "NR_20"){ return("red") }
        if (motif == "AP1_1"){ return("blue") }
        return("grey50")
    })
    return(res)
}
```

```r
chroms = c("chr20", "chr21", "chr22", "chrX")
lst_dat = lapply(chroms, function(chrom){
    fdiry  = file.path(FD_RES, "model_linear", "marginal_filter_input20x", chrom)
    fnames = dir(fdiry)
    
    mtfs = sapply(fnames, function(fname){
        mtf = fname
        mtf = str_remove(mtf, pattern = "motif_")
        mtf = str_remove(mtf, pattern = ".RDS")
        return(mtf)
    })
    
    lst_res = lapply(dir(fdiry), function(fname){
        fpath = file.path(fdiry, fname)
        lst   = readRDS(fpath)
    
        y = lst$y
        X = lst$X
    
        fit = lm(y ~ X + 0)
        res = summary(fit)
        res = stripGlmLR(res)
        return(res)
    })
    names(lst_res) = mtfs
    
    lst = lapply(names(lst_res), function(mtf){
        res = lst_res[[mtf]]
        dat = as.data.frame(coef(res))
        dat = rownames_to_column(dat, var="X")
        dat$Motif = mtf
        return(dat)
    })
    
    ###
    dat = bind_rows(lst)
    dat = dat %>% dplyr::filter(str_detect(X, "Intercept", negate = TRUE))
    dat$Color = fun(dat$Motif)
    dat$X = factor(dat$X, levels=c("XXTFX_DMSO", "XXTFX_Dex"))
    return(dat)
})
```

```r
lst_dat = lapply(lst_dat, function(dat){
    dat$adj_pval = p.adjust(dat$`Pr(>|t|)`, method = "BH")
    return(dat)
})
```

```r
lst = lst_dat
lst = lapply(lst, function(dat){
    dat = dat %>% 
        dplyr::select(Motif, X, Estimate, `Std. Error`, `t value`, `Pr(>|t|)`, adj_pval) %>%
        dplyr::rename(Sample          = X) %>%    
        dplyr::rename(Raw_pvalue      = `Pr(>|t|)`) %>%
        dplyr::rename(Adjusted_pvalue = adj_pval) %>%
        na.omit
    return(dat)
})
head(lst[[1]])
```

```r
lst_gpt = lapply(lst_dat, function(dat){
    gpt = ggplot(dat, aes(x=Estimate, y=-log10(adj_pval))) + 
        geom_point(size=0.7, color=dat$Color) + 
        geom_hline(yintercept = 2, color="red", alpha=0.5) +
        labs(x="Estimate", y="-log10(adj. p-value (BH))") +
        theme_bw() + 
        facet_wrap(~X)
    return(gpt)
})
lst_gpt[[1]] = lst_gpt[[1]] + ggtitle("Chr20")
lst_gpt[[2]] = lst_gpt[[2]] + ggtitle("Chr21")
lst_gpt[[3]] = lst_gpt[[3]] + ggtitle("Chr22")
lst_gpt[[4]] = lst_gpt[[4]] + ggtitle("ChrX")
```
