## Import merged counts

```r
### set column names and types
cnames = c("Chrom", "Start", "End", "Count")
ctypes = c(col_character(), col_double(), col_double(), col_double())
    
### import from input
fpath = file.path(FD_DAT, "Input/cnt_PER1.bed")
dat_cnt_input_per1 = read_tsv(
    fpath, 
    col_names = cnames, 
    col_types = ctypes)

fpath = file.path(FD_DAT, "Input/cnt_chr17.bed")
dat_cnt_input_chr17 = read_tsv(
    fpath, 
    col_names = cnames, 
    col_types = ctypes)

### import from output (DMSO)
fpath = file.path(FD_DAT, "TFX_DMSO/cnt_PER1.bed")
dat_cnt_dmso_per1 = read_tsv(
    fpath, 
    col_names = cnames, 
    col_types = ctypes)

fpath = file.path(FD_DAT, "TFX_DMSO/cnt_chr17.bed")
dat_cnt_dmso_chr17 = read_tsv(
    fpath, 
    col_names = cnames,
    col_types = ctypes)

### import from output (Dex)
fpath = file.path(FD_DAT, "TFX_Dex/cnt_PER1.bed")
dat_cnt_dex_per1 = read_tsv(
    fpath, 
    col_names = cnames, 
    col_types = ctypes)

fpath = file.path(FD_DAT, "TFX_Dex/cnt_chr17.bed")
dat_cnt_dex_chr17 = read_tsv(
    fpath, 
    col_names = cnames, 
    col_types = ctypes)
```

## Analysis: Overlapping of fragments

```r
df1 = dat_cnt_input_per1
df2 = dat_cnt_dmso_per1
dat = full_join(df1, df2, by=c("Chrom", "Start", "End"))

res = c(sum(df1$Count), sum(df2$Count))
names(res) = c("DF1", "DF2")
cat("\n#Total Fragments\n")
print(res)

res = c(nrow(df1), nrow(df2), nrow(dat), nrow(dat %>% na.omit))
names(res) = c("DF1", "DF2", "Full", "Overlap")
cat("\n#Total Unique Fragments\n")
print(res)

res = c(nrow(dat %>% na.omit)/nrow(df1),
        nrow(dat %>% na.omit)/nrow(df2))
names(res) = c("Overlap/DF1%", "Overlap/DF2%")
cat("\n#Total Unique Fragments (%)\n")
print(res)
```

```r
df1 = dat_cnt_input_chr17
df2 = dat_cnt_dmso_chr17
dat = full_join(df1, df2, by=c("Chrom", "Start", "End"))

res = c(sum(df1$Count), sum(df2$Count))
names(res) = c("DF1", "DF2")
cat("\n#Total Fragments\n")
print(res)

res = c(nrow(df1), nrow(df2), nrow(dat), nrow(dat %>% na.omit))
names(res) = c("DF1", "DF2", "Full", "Overlap")
cat("\n#Total Unique Fragments\n")
print(res)

res = c(nrow(dat %>% na.omit)/nrow(df1),
        nrow(dat %>% na.omit)/nrow(df2))
names(res) = c("Overlap/DF1%", "Overlap/DF2%")
cat("\n#Total Unique Fragments (%)\n")
print(res)
```

## Visualization: fragment coverage (PER1)

```r
### combine each sample
lst_dat = list(
    dat_cnt_input_per1,
    dat_cnt_dmso_per1,
    dat_cnt_dex_per1)
lst_name = c(
    "(Input; PER1)", 
    "(Output; DMSO; PER1)", 
    "(Output; Dex; PER1)")

### preprocess
lst_dat = lapply(lst_dat, function(dat){
    dat = dat %>% mutate(idx=1:nrow(dat)) %>% mutate(d=End-Start)
    return(dat)
})

### plot
lst_gpt = lapply(seq_along(lst_dat), function(idx){
    ### extract correspond dataframe and name
    dat = lst_dat[[idx]]
    idn = lst_name[idx]
    
    ### create ggplot
    gpt = ggplot(dat, aes(x=Start, xend=End, y=idx, yend=idx)) 
    gpt = gpt + 
        geom_segment() + 
        labs(
            x="Genome (chr17; hg38)", 
            y="Fragment Index", 
            title=paste("Unique STARR-seq fragment", idn, "sorted by start position")
        ) +
    theme_bw() +
    theme(
        title      = element_text(size=12),
        axis.title = element_text(size=12),
        axis.text  = element_text(size=12)
    )
    return(gpt)
})
```

```r
options(repr.plot.height=12, repr.plot.width=8)
grid.arrange(grobs=lst_gpt)
```
