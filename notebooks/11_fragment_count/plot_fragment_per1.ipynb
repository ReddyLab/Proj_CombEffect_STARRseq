```r
ctypes = c(
    col_character(), col_integer(), col_integer(), col_integer(),
    col_double(),  col_double(),
    col_integer(), col_integer(), col_integer(), col_integer(), 
    col_integer(), col_integer(), col_integer())
cnames = c(
    "Chrom", "Start", "End", "Count", 
    "AT", "GC", 
    "A", "C", "G", "T", 
    "N", "O", "len")
target = "target_PER1_annot.bed"

lst_dat = lapply(idn_samples, function(idn_sample){
    fpath = file.path(FD_WRK, "count_fragment", idn_sample, target)
    dat   = read_tsv(fpath, col_types=ctypes, col_names=cnames, comment = "#")
    return(dat)
})
```
