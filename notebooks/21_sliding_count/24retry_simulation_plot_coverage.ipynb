```r
fdiry = file.path(FD_RES, "sliding_window", "sliding_chr17_len_1kb_gap_100bp", "simulation")
fname = "dat_sim_input_20x_coverage_sim*.bed.gz"
fglob = file.path(fdiry, fname)

fpaths = Sys.glob(fglob)
fpaths
```

```r
cname = c("Chrom", "Start", "End", "Count")
ctype = c(col_character(), col_integer(), col_integer(), col_integer())

lst = lapply(seq_along(fpaths), function(idx){
    fpath = fpaths[idx]
    Sample = paste0("Input_20x_sim", idx)
    dat = read_tsv(fpath, col_types = ctype, col_names = cname)
    dat$Sample = Sample
    return(dat)
})

dat = bind_rows(lst)

dat_cov_sim_inp = dat
print(dim(dat))
head(dat)
```

```r
fdiry = file.path(FD_RES, "sliding_window", "sliding_chr17_len_1kb_gap_100bp", "simulation")
fname = "dat_sim_dmso_coverage_sim*.bed.gz"
fglob = file.path(fdiry, fname)

fpaths = Sys.glob(fglob)
fpaths
```

```r
fdiry = file.path(FD_RES, "sliding_window", "sliding_chr17_len_1kb_gap_100bp", "simulation")
fname = "dat_sim_dmso_coverage_sim*.bed.gz"
fglob = file.path(fdiry, fname)
fpaths = Sys.glob(fglob)

cname = c("Chrom", "Start", "End", "Count")
ctype = c(col_character(), col_integer(), col_integer(), col_integer())

lst = lapply(seq_along(fpaths), function(idx){
    fpath = fpaths[idx]
    Sample = paste0("TFX_DMSO_sim", idx)
    dat = read_tsv(fpath, col_types = ctype, col_names = cname)
    dat$Sample = Sample
    return(dat)
})

dat = bind_rows(lst)

dat_cov_sim_dmso = dat
print(dim(dat))
head(dat)
```

## summarize

```r
### calculate stats for each location
dat = dat_cov_sim_inp
dat = dat %>% 
    group_by(Chrom, Start, End) %>% 
    summarise(
        Cnt_Mean = mean(Count),
        Cnt_Std  = sd(Count),
        .groups = "drop") %>%
    mutate(
        Cnt_Up = Cnt_Mean + Cnt_Std,
        Cnt_Dw = Cnt_Mean - Cnt_Std)

### assign and show
dat_sum_sim_inp = dat
print(dim(dat))
head(dat)
```

```r
### calculate stats for each location
dat = dat_cov_sim_dmso
dat = dat %>% 
    group_by(Chrom, Start, End) %>% 
    summarise(
        Cnt_Mean = mean(Count),
        Cnt_Std  = sd(Count),
        .groups = "drop") %>%
    mutate(
        Cnt_Up = Cnt_Mean + Cnt_Std,
        Cnt_Dw = Cnt_Mean - Cnt_Std)

### assign and show
dat_sum_sim_dmso = dat
print(dim(dat))
head(dat)
```

```r
###
dat_cov = dat_cov_sim_inp %>% mutate(Loc = Start)
dat_sum = dat_sum_sim_inp %>% mutate(Loc = Start)

###
df1 = dat_cov
df2 = dat_sum


###
gp1 = ggplot(df1, aes(x=Loc, y=Count, color=Sample)) + 
    geom_line() + 
    theme_bw()  +
    labs(title="Chr17:8148000-8160000 (f00; sim)", x="Location", y="Count")

gp2 = ggplot(df2, aes(x=Loc, y=Cnt_Mean)) + 
    geom_ribbon(aes(ymin=Cnt_Dw, ymax=Cnt_Up), alpha=0.2) +
    geom_line(size=0.5) +
    theme_bw()  +
    labs(title="Chr17:8148000-8160000 (f00; sim)", x="Location", y="Count")

#gp1 = 
#leg = as._ggplot(get_legend(gp1))

options(repr.plot.height=6, repr.plot.width=7)
plot_grid(gp1, gp2, ncol = 1, align = "v")
```
