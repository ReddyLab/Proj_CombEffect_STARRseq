```r
GROUPS  = c(
    rep("Input",     5),
    rep("Input_20x", 5),
    rep("TFX_DMSO",  4),
    rep("TFX_Dex",   4))

SAMPLES = c(
    paste0("Input", 1:5),
    paste0("Input", 1:5, "_20x"),
    paste0("TFX", 2:5, "_DMSO"),
    paste0("TFX", 2:5, "_Dex"))
```

```r
fname = "target_PER1_count.bed.gz"
cname = c("Chrom", "Start", "End", "Count")
ctype = c(col_character(), col_integer(), col_integer(), col_integer())

lst = lapply(SAMPLES, function(SAMPLE){
    
    ### set file path
    fdiry = file.path(FD_RES, "sliding_window", "sliding_PER1_len_1kb_gap_100bp", SAMPLE)     
    fpath = file.path(fdiry, fname)
    
    ### import data
    dat = read_tsv(fpath, col_types = ctype, col_names = cname)
    dat$Sample = SAMPLE
    return(dat)
})

####
dat = bind_rows(lst)
dat = left_join(dat, dat_lib, by="Sample")
dat = dat %>% mutate(CPM = Count * 10^6 / Size)
#dat = dat %>% spread(Sample, Count)

###
dat_win_cnt = dat
print(dim(dat))
head(dat)
```

## Count Distribution

```r
dat = dat_win_cnt
gpt = ggplot(dat, aes(x=Count)) + 
    geom_histogram(bins=30) + 
    facet_wrap(~Sample, scale="free") + 
    theme_cowplot() + labs(x="Count of overlapped fragment in each sliding window")

options(repr.plot.height=7, repr.plot.width=15)
print(gpt)
```

## Mean-Variance Relationship

```r
dat = dat_win_cnt
dat = dat %>% 
    dplyr::group_by(Chrom, Start, End, Group) %>% 
    dplyr::summarise(Mean = mean(CPM), Var = var(CPM), .groups = "drop")


dat_mu_var = dat
print(dim(dat))
head(dat)
```

```r
### try calculation similar to voom
dat = dat_win_cnt
dat = dat %>% 
    dplyr::group_by(Chrom, Start, End, Group) %>% 
    dplyr::summarise(Mean = mean(CPM), Var = var(CPM), .groups = "drop")


dat_mu_var = dat
print(dim(dat))
head(dat)
```

```r
dat = dat_mu_var
gpt = ggplot(dat, aes(x=Mean, y=Var)) + 
    geom_point(size=0.5) + 
    theme_cowplot() + 
    facet_wrap(~Group, scale="free") +
    labs(x="Mean", y="Variance", title="Mean-Variance Relationship of CPM")

options(repr.plot.height=7, repr.plot.width=7)
print(gpt)
```

```r
dat = dat_mu_var
gpt = ggplot(dat, aes(x=Mean, y=Var)) + 
    geom_point(size=0.5) + 
    scale_x_log10() + 
    scale_y_log10() +
    theme_cowplot() + 
    facet_wrap(~Group, scale="free") + 
    labs(x="Mean", y="Variance", title="Mean-Variance Relationship of CPM (Log-Log)")

options(repr.plot.height=7, repr.plot.width=7)
print(gpt)
```
