```r
CHROMS = paste0("chr", c(1:22, "X", "Y"))
CHROMS
```

```r
fdiry = file.path(FD_RES, "count_fragment")
cnames = c("Sample", "Chrom", "Count")

lst_dat = lapply(CHROMS, function(CHROM){
    fname = paste0("count_unique_fragment_", CHROM, ".txt")
    fpath = file.path(fdiry, fname)
    dat   = read_tsv(fpath, col_names = cnames, show_col_types = FALSE)
    return(dat)
})

dat_cnt = bind_rows(lst_dat)
dat_cnt
```

```r
### Helper function to get
get_sample = function(idn_sample){
    idn = idn_sample
    idn = str_replace(
        string = idn, 
        pattern = "Input[0-9]", 
        replacement = "Input")
    idn = str_replace(
        string = idn, 
        pattern = "TFX[0-9]_", 
        replacement="TFX_")
    return(idn)
}
```

```r
dat = dat_cnt
dat = dat %>% 
    mutate(Group = get_sample(Sample)) %>% 
    group_by(Group) %>%
    summarize(Count = sum(Count)/10^6, .groups = 'drop')
dat
```
