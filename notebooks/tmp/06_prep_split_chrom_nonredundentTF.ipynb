[combining two magics in jupyter](https://stackoverflow.com/questions/53204167/is-it-possible-to-combine-magics-in-ipython-jupyter)

````python
from IPython.core.magic import register_cell_magic

@register_cell_magic
def accio(line, cell):
    ipy = get_ipython()
    expr = ipy.input_transformer_manager.transform_cell(cell)
    expr_ast = ipy.compile.ast_parse(expr)
    expr_ast = ipy.transform_ast(expr_ast)
    code = ipy.compile(expr_ast, '', 'exec')
    exec(code)
````
