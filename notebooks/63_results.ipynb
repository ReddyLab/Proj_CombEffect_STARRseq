```python
import numpy as np
import arviz as az
import os, json
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fdiry = "./model_kmer_sim/"

### get input
fname = "input_trivial_v1.json"
with open(os.path.join(fdiry, fname), "r") as file:
    dat_input = json.load(file)
    
### get output
fname = "output_trivial_v1.csv"
dat_output = az.from_cmdstan(posterior=os.path.join(fdiry, fname))
dat_output
```

```python
# traceplot
az.plot_trace(dat_output)
plt.show()
```

```python
dat = dat_output
seg = dat.posterior.data_vars.get("SEG").to_series()
seg_true = dat_input["SEG_TRUE"]

print("TRUE  MEDIAN  MEAN  VAR")
for i in range(15):
    x = seg.loc[0,:,i]
    print("{:.2f}  {:.2f}    {:.2f}  {:.2f}".format(
        seg_true[i], 
        np.median(x), 
        np.mean(x), 
        np.var(x)))
```
